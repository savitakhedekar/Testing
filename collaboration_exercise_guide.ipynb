
# Step 1: Set Up the Repository

## Create a New Repository:

- Go to GitHub and sign in.
- Click on the "+" icon in the top right corner and select "New repository."
- Name your repository `collaboration-exercise`.
- Add a description (optional) and select "Public."
- Check the box to "Initialize this repository with a README."
- Click "Create repository."

## Clone the Repository Using GitHub Desktop:

- Open GitHub Desktop.
- Click "File" > "Clone Repository."
- Select the `collaboration-exercise` repository and click "Clone."
    


# Step 2: Make Initial Commit

## Create a New File:

- Open the cloned repository in your code editor.
- Create a new file named `team.txt`.
- Add the following content:

```markdown
# Team Members

- [Your Name]
```

## Commit and Push:

- Save the file.
- Go to GitHub Desktop.
- You should see `team.txt` under "Changes."
- Write a commit message: "Add initial team member."
- Click "Commit to main."
- Click "Push origin" to push the changes to GitHub.



# Step 3: Collaborators Fork and Clone the Repository

## Fork the Repository:

- Collaborators go to the original repository on GitHub.
- Click the "Fork" button in the top right corner.

## Clone the Forked Repository Using GitHub Desktop:

- Open GitHub Desktop.
- Click "File" > "Clone Repository."
- Select the forked repository and click "Clone."



# Step 4: Collaborator Makes Changes

## Edit `team.txt`:

- The collaborator opens the `team.txt` file in their code editor.
- They add their name to the file:

```markdown
# Team Members

- [Your Name]
- [Collaborator's Name]
```

## Commit and Push:

- The collaborator saves the file.
- In GitHub Desktop, they commit the changes with the message "Add collaborator's name."
- They push the changes to their fork on GitHub.

## Create a Pull Request:

- On GitHub, the collaborator creates a pull request to merge their changes into the original repository.



# Step 5: Create a Merge Conflict

## Create a New Branch:

- In GitHub Desktop, create a new branch from the main branch named `add-feature`.
- Switch to the `add-feature` branch.

## Edit `team.txt`:

- In the `add-feature` branch, open the `team.txt` file.
- Add a new line with your name and a feature:

```markdown
# Team Members

- [Your Name]
- [Collaborator's Name]

## Feature
- [Your Name]: Working on feature X
```

## Commit and Push:

- Save the file.
- Commit the changes in GitHub Desktop with the message "Add feature section."
- Push the changes to GitHub.

## Simultaneously Edit `team.txt` in Main Branch:

- Switch back to the main branch in GitHub Desktop.
- Edit the `team.txt` file:

```markdown
# Team Members

- [Your Name]
- [Collaborator's Name]

## Tasks
- [Your Name]: Working on task Y
```

## Commit and Push:

- Save the file.
- Commit the changes in GitHub Desktop with the message "Add tasks section."
- Push the changes to GitHub.



# Step 6: Merge and Resolve Conflicts

## Create a Pull Request:

- On GitHub, create a pull request to merge the `add-feature` branch into the main branch.

## Identify Merge Conflict:

- GitHub will indicate a merge conflict.

## Resolve Conflict Using GitHub Desktop:

- In GitHub Desktop, fetch the latest changes.
- Switch to the main branch.
- Click "Branch" > "Merge into Current Branch" > select `add-feature`.
- GitHub Desktop will show the conflict.

## Resolve Conflict:

- Open the conflicting file (`team.txt`) in your code editor.
- You will see conflict markers like:

```markdown
<<<<<<< main
- [Your Name]: Working on task Y
=======
- [Your Name]: Working on feature X
>>>>>>> add-feature
```

- Resolve the conflict by editing the file to combine the changes:

```markdown
# Team Members

- [Your Name]
- [Collaborator's Name]

## Feature
- [Your Name]: Working on feature X

## Tasks
- [Your Name]: Working on task Y
```

## Commit and Push Resolved Changes:

- Save the resolved file.
- In GitHub Desktop, commit the merge with a message like "Resolve merge conflict in team.txt."
- Push the changes to GitHub.



# Step 7: Verify the Merge

## Merge Pull Request:

- On GitHub, merge the pull request if it is not automatically merged after resolving conflicts.

## Pull Latest Changes:

- Collaborators should pull the latest changes to ensure everyone is up to date.
